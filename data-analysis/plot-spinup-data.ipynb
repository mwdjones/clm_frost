```python
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
import cftime
```

```python
save_dir = '/glade/u/home/marielj/clm_frost/data-analysis/figures/spinup-data/'

#import full transient spinup data
spinup = xr.open_mfdataset('/glade/scratch/marielj/1ptBGC_spinup/run/1ptBGC_spinup.clm2.h0.*.nc', 
                          parallel = True)
```

```python
def plot_spinup(var):
    fig, ax = plt.subplots(figsize = (6,4))

    #Plot accelerated spinup data
    spinup[var].plot(linewidth = 1, alpha = 0.4, ax = ax, color = 'red')

    #Plot rolling average
    roll = spinup[var].rolling(time = 30).mean()
    plt.plot(spinup.time, roll, color = 'red', linewidth = 0.5)
    
    ax.set_title(spinup[var].long_name)

    #plt.legend()
    plt.savefig(save_dir + str(var) + '_forcings.pdf', bbox_inches = 'tight')
    plt.show() 
```

```python
spinup_vars = ['GPP', 'NPP', 'TLAI',
               'TOTECOSYSC', 'TOTECOSYSN', 'TOTSOMC', 'TOTSOMN', 'TOTVEGC', 'TOTVEGN', 'TWS']

for v in spinup_vars:
    plot_spinup(v)
```
